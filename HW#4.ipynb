### Оценка качества исправления ошибок

Описание метода: для каждой позиции референса записываем нуклеотиды из всех ридов, которые покрывают данную позицию; если данная позиция референса не покрыта ридами, то записываем 0. Далее считаем консенсус по каждой покрытой позиции референса - так считаем ошибки / совпадения в сырых и исправленных ридах. Удалённые ошибки / совпадения из сырых ридов считаем, смотря на не покрытые позиции референса в исправленных ридах. Статистика выводится в процентах.

```python
# читаем референс
ref_file = open('MG1655-K12.first400K.fasta', 'r')
ref_length = 0
ref_seq = ''
ref_file.readline()
for line in ref_file:
    line = line.strip()
    ref_length += len(line)
    ref_seq += line
```

```python
def get_coverage_letters(sam_file_name, ref_seq, ref_length):    
    sam_file = open(sam_file_name, 'r')
    sam_file.readline()
    sam_file.readline()
    per_base_letters = [0]* ref_length
    for line in sam_file:
        line = line.strip().split()
        start = int(line[3]) - 1
        if start == -1: # unmapped read
            continue
        else:
            read_length = len(line[9])
            read_seq = line[9]
            stop = min(start + read_length - 1, ref_length)
            for pos in range(start, stop):
                if per_base_letters[pos] == 0:
                    per_base_letters[pos] = [list(read_seq)[pos - start]]
                else:
                    per_base_letters[pos].append(list(read_seq)[pos - start])
    ref_file.close()
    sam_file.close()
    return per_base_letters
```

```python
coverage_letters_raw = get_coverage_letters('alignment_raw.sam', ref_seq, ref_length)
coverage_letters_trim = get_coverage_letters('alignment_trim.sam', ref_seq, ref_length)
coverage_letters_bh = get_coverage_letters('alignment_bh.sam', ref_seq, ref_length)
```

```python
def get_correction_estimation(per_base_raw, per_base_corr, ref_seq):
    data = np.zeros((2, 3))
    for i in range(len(per_base_corr)):
        if per_base_corr[i] == 0:
            if per_base_raw[i] != 0:
                counter = Counter(per_base_raw[i])
                most_common = counter.most_common(1)[0][0]
                if most_common != ref_seq[i] and most_common != 'N':
                    data[0, 2] += 1
                elif most_common == ref_seq[i] and most_common != 'N':
                    data[1, 2] += 1
        elif per_base_corr[i] != 0:
            if per_base_raw[i] != 0:
                counter_raw = Counter(per_base_raw[i])
                counter_corr = Counter(per_base_corr[i])
                most_common_raw = counter_raw.most_common(1)[0][0]
                most_common_corr = counter_corr.most_common(1)[0][0]
                if most_common_raw != 'N' and most_common_corr == 'N':
                    if most_common_raw != ref_seq[i]:
                        data[0, 2] += 1
                    else:
                        data[1, 2] += 1
                elif most_common_raw != 'N' and most_common_corr != 'N':
                    if most_common_raw != ref_seq[i] and most_common_corr != ref_seq[i]:
                        data[0, 0] += 1
                    elif most_common_raw != ref_seq[i] and most_common_corr == ref_seq[i]:
                        data[0, 1] += 1
                    elif most_common_raw == ref_seq[i] and most_common_corr != ref_seq[i]:
                        data[1, 0] += 1
                    elif most_common_raw == ref_seq[i] and most_common_corr == ref_seq[i]:
                        data[1, 1] += 1
    return (data / len(ref_seq)) * 100
```

Данные trimmomatic:

```python
data_trim = get_correction_estimation(coverage_letters_raw, coverage_letters_trim, ref_seq)
data_trim_df = pd.DataFrame(data_trim, columns = ['Error corr reads', 'Correct corr reads', 'Abscent corr reads'], index = ['Error raw data', 'Correct raw data'])
print(data_trim_df)
```

Видим, что процент незамеченных ошибок намного выше процента замеченных и исправленных ошибок. Исправленных правильных баз нет, все правильные нуклеотиды остались, что хорошо. Возможно, это связано с тем, что trimmomatic не исправляет ошибки, а удаляет риды и отдельные базы, из-за чего может "смещаться" консенсус. Возможно также, что нужно подобрать другие параметры и получить другие исправленные риды.

Данные BayesHammer:

```python
data_bh = get_correction_estimation(coverage_letters_raw, coverage_letters_bh, ref_seq)
data_bh_df = pd.DataFrame(data_bh, columns = ['Error corr reads', 'Correct corr reads', 'Abscent corr reads'], index = ['Error raw data', 'Correct raw data'])
print(data_bh_df)
```

Видим, что исправленных ошибок в итоге не оказалось, наоборот, примерно полпроцента правильных баз были исправлены на неправильные. Все ошибки оказались незамеченными. Возможно, здесь тоже всё дело в параметрах программы.
